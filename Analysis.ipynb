# Analyses

```python
wealth = pd.read_csv('./data/wealth-track.csv')

wealth = wealth[['rich-artist','colonizer', 'poor-artist', 'poor-phillistine', 'rich-phillistine']]
#wealth = wealth.drop(wealth.index[[0]])
```

```python
wealth = wealth.astype('float')
wealth
```

## change in wealth of tracked individuals
Looks like poor of all kinds have fewer mobility as indicated by the lack in change in neighborhood wealth over time. The colonizer has choice.

```python
change = wealth[['colonizer', 'poor-artist', 'poor-phillistine']].pct_change(periods=400)
change.plot()
plt.title('Percent Change in neighborhood wealth')
plt.ylabel('% Change')
plt.xlabel('Time')
plt.show()
```

```python
change = change.loc[450:]
means = dict({
    'mean': [np.mean(change[['colonizer']]),np.mean(change[['poor-artist']]),np.mean(change[['poor-phillistine']])],
    'std' : [np.std(change[['colonizer']]),np.std(change[['poor-artist']]),np.std(change[['poor-phillistine']])]
})

means = pd.DataFrame.from_dict(means).astype('float')
means.index = ['Colonizer', 'Poor-Artist', 'Poor-Phillistine']
```

```python
gini = pd.read_csv('./data/gini-test-clean.csv')
gini = gini[['"neighborhood 0"', '"neighborhood 1"', '"neighborhood 2"', '"neighborhood 3"', '"neighborhood 4"' ]]
gini = gini.astype('float')
```

```python
gini_change = gini.pct_change(periods = 200)
gini_change.loc[2000:4000].plot()
plt.show()
```

```python
art = pd.read_csv('./data/art-test.csv')
art = art[['"neighborhood 0"', '"neighborhood 1"', '"neighborhood 2"', '"neighborhood 3"', '"neighborhood 4"' ]]
art_change = art.pct_change(periods = 200)
```

```python
wealth = pd.read_csv('./data/wealth-test.csv')
wealth = wealth[['"neighborhood 0"', '"neighborhood 1"', '"neighborhood 2"', '"neighborhood 3"', '"neighborhood 4"' ]]
wealth_change = wealth.pct_change(periods = 200)
```

# Change in Gini is correlated with change in art positively and negatively with wealth
which is to say that when there is a small amount of wealth in a neighbohood, then it is controlled by a small amount of people

```python
ar_cor = float(np.mean(art_change.corrwith(wealth_change, axis = 1, drop=True)))
g_cor_w = np.mean(wealth_change.corrwith(gini_change, axis = 1, drop = True))
g_cor_a = np.mean(art_change.corrwith(gini_change, axis = 1, drop = True))
print(ar_cor)
print(g_cor_w)
print(g_cor_a)
```
